```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

data = pd.read_csv("/content/po2_data.csv")
data
```

```python
jitter_columns = ['jitter(%)', 'jitter(abs)', 'jitter(rap)', 'jitter(ppq5)', 'jitter(ddp)']

data['jitter_mean'] = data[jitter_columns].mean(axis=1)
data.head()
```

```python
shimmer_columns = ['shimmer(%)', 'shimmer(abs)', 'shimmer(apq3)','shimmer(apq5)', 'shimmer(apq11)', 'shimmer(dda)']

data['shimmer_mean'] = data[shimmer_columns].mean(axis=1)
data.head()
```

```python
# Checking skewness:

from scipy import stats

var = ['age', 'nhr', 'hnr', 'rpde', 'dfa', 'ppe', 'jitter_mean', 'shimmer_mean']

for i in var:
  skewness = stats.skew(data[i])
  print(f"Skewness of {i}: ", skewness)


```

```python
# Log_transformation

r_skewed = ['nhr','jitter_mean', 'shimmer_mean']

for j in r_skewed:
  data[f'log_{j}'] = np.log(data[j])
  skewness = stats.skew(data[f"log_{j}"])
  print(f"Skewness of log_{j}: ", skewness)
```

```python
columns =  ['nhr','shimmer(%)', 'shimmer(abs)', 'shimmer(apq3)','shimmer(apq5)', 'shimmer(apq11)', 'shimmer(dda)', 'jitter(%)', 'jitter(abs)', 'jitter(rap)', 'jitter(ppq5)', 'jitter(ddp)', 'jitter_mean','shimmer_mean']
data.drop(columns, inplace = True, axis=1)
data
```

```python
data.columns
```

```python
from sklearn.preprocessing import StandardScaler


continuous_columns = ['test_time', 'hnr', 'rpde', 'dfa', 'ppe', 'log_nhr' ,'log_jitter_mean', 'log_shimmer_mean']
scaler = StandardScaler()
data[continuous_columns] = scaler.fit_transform(data[continuous_columns])
data[continuous_columns].describe()
```

```python
from sklearn.preprocessing import PowerTransformer

continuous_columns = ['test_time', 'log_nhr', 'hnr', 'rpde', 'dfa', 'ppe', 'log_jitter_mean', 'log_shimmer_mean']
transformer = PowerTransformer(method='yeo-johnson')
data[continuous_columns] = transformer.fit_transform(data[continuous_columns])
```

```python
from scipy import stats

var = ['test_time','log_nhr', 'hnr', 'rpde', 'dfa', 'ppe', 'log_jitter_mean', 'log_shimmer_mean']

for i in var:
  skewness = stats.skew(data[i])
  print(f"Skewness of {i}: ", skewness)
```

```python
from sklearn.model_selection import train_test_split

X = ['age', 'log_nhr', 'hnr', 'rpde', 'dfa', 'ppe', 'log_jitter_mean', 'log_shimmer_mean']
y = ['motor_updrs', 'total_updrs']

X_train, X_test, y_train, y_test = train_test_split(data[X], data[y], test_size=0.2, random_state=42)

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)




```

```python
y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
nrmse = rmse / np.abs(np.mean(y_test))
r2 = r2_score(y_test, y_pred)

n = len(y_test)
p = len(X)

adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("NRMSE:", nrmse)
print("r2:", r2)
print("Adjusted-r2:", adjusted_r2)
```
